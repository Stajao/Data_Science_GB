```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
boston = load_boston()
```

```python
type(boston)
```

```python
data = boston['data']
data[0:3]
```

```python
feature_names = boston['feature_names']
feature_names
```

```python
target = boston['target']
target[0:15]
```

```python
x = pd.DataFrame(data, columns=feature_names)
x.head()
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred[:10]
```

```python
check = pd.DataFrame({'y_test':y_test['price'], 'y_pred':y_pred.flatten()})
check
```

```python
check['error'] = check['y_pred'] - check['y_test']
check.head()
```

```python
mse1 = (check["error"] ** 2).mean()
mse1
```

```python
(np.abs(check["error"])).mean()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(check['y_test'], check['y_pred'])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
DATASET_PATH = './creditcard.csv'
df = pd.read_csv(DATASET_PATH, sep=',')
df.head(10)
```

```python
y = df['Class']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
mdl = GridSearchCV(estimator=RandomForestClassifier(random_state=100),param_grid=parameters,scoring='roc_auc',cv=3)
```

```python
mdl.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(mdl.cv_results_)
cv_results.head()
```

```python
mdl.best_params_
```

```python
mdl = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15)
```

```python
mdl.fit(X_train, y_train)
```

```python
y_pred_proba = mdl.predict_proba(X_test)[:, 1]
```

```python
y_pred_proba[:5]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
